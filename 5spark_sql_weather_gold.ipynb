```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *

spark = SparkSession.builder.master('local[2]').getOrCreate()
```

```python
# Create a DataFrame from parquet data (automatically infer schema and data types)
# There are other file formats you can read from (e.g., csv, orc, parquet)
# https://spark.apache.org/docs/2.2.0/sql-programming-guide.html#data-sources

weatherDFSilverFilePath = "hdfs://hdfs-nn:9000/warehouse/tabd.db/parquet_table_weather/"

# Read Parquet table fromo silver weather data
parqWeatherDFSilver = spark.read.parquet(weatherDFSilverFilePath)

```

```python
parqWeatherDFSilver.printSchema()
parqWeatherDFSilver.show()
parqWeatherDFSilver.toPandas()
```

```python
from pyspark.sql.types import IntegerType
```

```python
DF1= parqWeatherDFSilver.withColumn("day_of_year", date_format(col("date"), "D").cast(IntegerType()))
DF1.printSchema()
DF1.toPandas()
```

```python
# First you need to create a temporary view
DF1.createOrReplaceTempView("dayofyear")

# Then you can write all the SQL you want, as you're json is seen as a regular table now.
sqlized_df = spark.sql(
    """
    SELECT DISTINCT date, day_of_year
    FROM dayofyear
    WHERE date = "2019-03-20" OR date = "2019-06-21" OR date = "2019-09-22" OR date = "2019-12-21" and station_id="1200545"
    ORDER BY date
    """
)

sqlized_df.toPandas()
```

```python
DF2 = DF1.withColumn("season", 
                     when((col("day_of_year") > 78) & (col("day_of_year") < 172), "Primavera") \
                     .when((col("day_of_year") > 171) & (col("day_of_year") < 265), "Verão") \
                     .when((col("day_of_year") > 264) & (col("day_of_year") < 355), "Outono") \
                     .otherwise("Inverno"))
                         

DF2.printSchema()
DF2.toPandas()
```

```python
# First you need to create a temporary view
DF2.createOrReplaceTempView("season")

# Then you can write all the SQL you want, as you're json is seen as a regular table now.
sqlized_df = spark.sql(
    """
    SELECT DISTINCT date, day_of_year, season
    FROM season
    WHERE date = "2019-03-20" OR date = "2019-06-21" OR date = "2019-09-22" OR date = "2019-12-21" and station_id="1200545"
    ORDER BY date
    """
)

sqlized_df.toPandas()
```

```python
DF3 = DF2.withColumn("country", lit("Portugal")) \
                                    .withColumn("city", 
                                                when((col("station_id") == "1200545"), "Porto") \
                                                .when((col("station_id") == "1200548"), "Coimbra") \
                                                .when((col("station_id") == "1200551"), "Viana do Castelo") \
                                                .when((col("station_id") == "1200554"), "Faro") \
                                                .when((col("station_id") == "1200558"), "Évora") \
                                                .when((col("station_id") == "1200560"), "Viseu") \
                                                .when((col("station_id") == "1200562"), "Beja") \
                                                .when((col("station_id") == "1200567"), "Vila Real") \
                                                .when((col("station_id") == "1200570"), "Castelo Branco") \
                                                .when((col("station_id") == "1200571"), "Portalegre") \
                                                .when((col("station_id") == "1200575"), "Bragança") \
                                                .when((col("station_id") == "1200579"), "Lisboa") \
                                                .when((col("station_id") == "1210622"), "Braga") \
                                                .when((col("station_id") == "1210683"), "Guarda") \
                                                .when((col("station_id") == "1210702"), "Aveiro") \
                                                .when((col("station_id") == "1210718"), "Leiria") \
                                                .when((col("station_id") == "1210734"), "Santarém") \
                                                .when((col("station_id") == "1210770"), "Setúbal") \
                                                .when((col("station_id").isNull()) ,"desconhecida") \
                                                .otherwise("desconhecida"))
                         

DF3.printSchema()
DF3.toPandas()
```

```python
DF4=DF3.withColumn("air_temperature_avg",
                         when((col("air_temperature_avg") == -990.0), None) \
                         .otherwise(col("air_temperature_avg")))

DF5=DF4.withColumn("air_temperature_min",
                         when((col("air_temperature_min") == -990.0), None) \
                         .otherwise(col("air_temperature_min")))

DF6=DF5.withColumn("air_temperature_max",
                         when((col("air_temperature_max") == -990.0), None) \
                         .otherwise(col("air_temperature_max")))

DF7=DF6.withColumn("global_radiation_total",
                         when((col("global_radiation_total") == -990.0), None) \
                         .otherwise(col("global_radiation_total")))

weatherDFGold=DF7.withColumn("rain_precipitation_qty",
                         when((col("rain_precipitation_qty") == -990.0), None) \
                         .otherwise(col("rain_precipitation_qty")))


weatherDFGold.printSchema()
weatherDFGold.toPandas()
```

```python
# First you need to create a temporary view
weatherDFGold.createOrReplaceTempView("weathergold")

# Then you can write all the SQL you want, as you're json is seen as a regular table now.
sqlized_df = spark.sql(
    """
    SELECT * 
    FROM weathergold
    WHERE year = 2020
    """
)

sqlized_df.toPandas()
```

```python
# with partitions
weatherDFGold \
    .repartition(1) \
    .write \
    .partitionBy("year") \
    .format("parquet") \
    .mode("overwrite") \
    .save("hdfs://hdfs-nn:9000/warehouse/tabd.db/parquet_table_weather_gold/")
```
